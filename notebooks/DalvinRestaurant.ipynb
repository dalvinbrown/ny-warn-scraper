```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/warn.csv')
```

```python
df
```

```python
df[df['business_type'] == 'Restaurant']
```

```python
restaurant = df[df['business_type'].str.contains("Restaurant")]
```

```python
restaurant
```

```python
layoff = restaurant['notice_date'].value_counts().reset_index().rename(columns = {'index': 'date', 'notice_date': 'count'})
layoff
```

```python
results = layoff['date'].sort_values(ascending=False).reset_index()
results
```

```python
results.to_csv('../output/results.csv', index=False)
```
